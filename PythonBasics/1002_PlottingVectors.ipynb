```
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-4,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vec(vecs):
  colors=['r','g','b']
  i=0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)])
    i+=1
  plt.xlim(-20,20)
  plt.ylim(-20,20)
  plt.show()
  
plot_vec([(0,0,3,4),(0,0,-3,8),(0,0,6,5)])
```

```
vecs=[np.asarray([0,0,3,4]),np.asarray([0,0,-3,8]),np.asarray([0,0,6,5])]
plot_vec([vecs[0],vecs[1],vecs[0]+vecs[1],vecs[0]-vecs[1],-vecs[0],-vecs[0]+vecs[1]])
```

```
a=np.asarray([3,4])
b=np.asarray([1,2])
```

$\vec{a}\cdot\vec{b}=|\vec{a}|\cdot|\vec{b}|\cos(\theta)=a_xb_x+a_yb_y$

```
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

$proj(a_b)=|\vec{a}|\cos(\theta)=|\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}|\cdot|\vec{b}|}=\frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```
a_b=a_dot_b/np.linalg.norm(b)
print(a_b)
```

$\vec{a_b}=a_b\hat{b}=a_b\frac{\vec{b}}{|\vec{b}|}$

```
vec_a_b=(a_b/np.linalg.norm(b))*b
print(vec_a_b)
```
