## Sets, Dictionaries

```
name = 'iPhone 5s'
is_android=False
screen_size=4.3
my_phone=[name,is_android,screen_size]
```

```
my_set = set(my_phone)
print(my_set)
```

```
my_phone.append(4.3)
print(my_phone)
```

```
#my_set.append(4.3) -> set can't be appended and has only unique values
my_new_set = set(my_phone)
print(my_new_set)
```

```
import time
```

```
my_list = list(range(1000000))
tic = time.time()
print(98590 in my_list)
print(1233404 in my_list)
print(10000001 in my_list)
toc = time.time()
print("elapsed time= ",toc-tic)
```

```
my_set = set(range(1000000))
tic = time.time()
print(98590 in my_set)
print(1233404 in my_set)
print(10000001 in my_set)
toc = time.time()
print("elapsed time= ",toc-tic)
```

Set operations are faster

## Python Classes

```
class MobilePhone:
  """This is a simple example for python class"""
  def __init__(self,name,is_android=False,screen_size=4.3):
    self.name = name
    self.is_android = is_android
    self.screen_size = screen_size
    self.rating = -1

  def has_rating(self):
    return self.rating > -1
```

```
new_phone = MobilePhone('iphone5s')
```

```
print(type(new_phone))
```

```
new_phone1 = MobilePhone('oneplus',True,7)
print(new_phone1.screen_size)
```

```
new_phone1.screen_size = 6.5
print(new_phone1.screen_size)
```

```
print(new_phone.has_rating())
```

```
new_phone1.rating =5
print(new_phone1.has_rating())
```

```
MobilePhone.__doc__
```

```
class iPhone(MobilePhone):
  def __init__(self,name):
    MobilePhone.__init__(self,name,False,4)
  def __str__(self):
    return self.name + " " +str(self.is_android)+" "  + str(self.screen_size)
```

```
new_phone2 = iPhone('iphone5s')
print(type(new_phone2))
```

```
print(new_phone2.rating)
```

```
print(new_phone2)
```
